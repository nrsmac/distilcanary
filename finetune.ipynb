```python
!pip install datasets
from datasets import load_dataset

dataset = load_dataset('allenai/prosocial-dialog')
dataset = dataset.class_encode_column('safety_label')
dataset = dataset.rename_column('safety_label', 'label')
dataset = dataset.rename_column('context', 'text')
dataset = dataset.remove_columns(['response', 'rots', 'safety_annotations', 'safety_annotation_reasons', 'source', 'etc', 'dialogue_id', 'response_id', 'episode_done'])
dataset["train"][100]
```

```python
!pip install transformers
from transformers import AutoTokenizer

model_name = 'distilbert-base-cased'
tokenizer = AutoTokenizer.from_pretrained(model_name)


def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True)


tokenized_datasets = dataset.map(tokenize_function, batched=True)

train_dataset = tokenized_datasets["train"].shuffle(seed=42)
eval_dataset = tokenized_datasets["test"].shuffle(seed=42)
```

```python
train_dataset
```

```python
from transformers import AutoModelForSequenceClassification

model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=5)
```

```python
from transformers import AutoModelForSequenceClassification
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=5)
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(output_dir="distillcanary")
```

```python
import numpy as np
!pip install evaluate
import evaluate

metric = evaluate.load("accuracy")
```

```python
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return metric.compute(predictions=predictions, references=labels)
```

```python
!pip install wandb
import wandb
wandb.login()
%env WANDB_PROJECT=distillbert_canary
```

```python
from transformers import TrainingArguments, Trainer
training_args = TrainingArguments(output_dir="distillcanary", report_to="wandb",do_eval=True, evaluation_strategy='epoch')
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
    compute_metrics=compute_metrics,
)
```

```python
trainer.train()
wandb.finish()
```
